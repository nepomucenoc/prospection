```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Projetos_ion/ION_Prospeccao_Inteligente/datasets/geolocalizacao/lat_lon_nominatim.csv', header=None)
df.head()
```

```python
df[1]
```

```python
lista_lat_long = []
for lat_long in df[1]:
  lista_lat_long.append(str(lat_long).split(','))
lista_lat_long
```

```python
latitude = []
for indice1 in range(len(lista_lat_long)):
  latitude.append(lista_lat_long[indice1][0])

latitude
```

```python
longitude = []
for indice1 in range(len(lista_lat_long)):
  longitude.append(lista_lat_long[indice1][1])

longitude
```

```python
df['latitude'] = latitude
df['longitude'] = longitude

df.head()
```

```python
df.columns = ['endereco', 'latlong', 'hash', 'latitude', 'longitude']
df.head()
```

```python
df = df[['endereco', 'latitude', 'longitude', 'hash']]
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/Projetos_ion/ION_Prospeccao_Inteligente/datasets/geolocalizacao/lat_lon_nominatim1.csv', header=None, index=False)
```
