```python
import pandas as pd

from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('/content/drive/MyDrive/Projetos_ion/ION_Prospeccao_Inteligente/datasets/geolocalizacao/enderecos_hash_no_duplicate_hash.csv')
df.head()
```

```python
df1 = df.head(20)
```

```python
df1.to_csv('df_teste1000.csv', index=False)
```

```python
locator = Nominatim(user_agent="myGeocoder")
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
df1['endereco_geolocalizado'] = df1['endereco_completo'].apply(geocode)
```

```python
df1['latitude'] = df1['endereco_geolocalizado'].apply(lambda loc: loc.latitude if loc else None)
df1['longitude'] = df1['endereco_geolocalizado'].apply(lambda loc: loc.longitude if loc else None)
df1['geocoder'] = df1['endereco_geolocalizado'].apply(lambda loc: tuple(loc.point) if loc else None)
df1.head()
```
