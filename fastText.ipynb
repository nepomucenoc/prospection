```python
def generateVector(sentence):
    return ft.get_sentence_vector(sentence)
```

```python
vector1 = generateVector('Comércio varejista de mercadorias em geral, com predominância de produtos alimentícios - minimercados, mercearias e armazéns')
vector1
```

```python
df = pd.DataFrame()
df['CNAE_descricao'] = lista
df.head()
```

```python
df['vector'] = df['CNAE_descricao'].apply(generateVector)
```

```python
df.head()
```

## Euclidean

```python
def spatialDistance_euclidean(vector1, vector2):
    return spatial.distance.euclidean(vector1, vector2)
```

```python
df['score_euclidean'] = df.apply(lambda x: spatialDistance_euclidean(vector1, x['vector']), axis=1)
df = df.drop_duplicates(subset=['score_euclidean']).sort_values(by=['score_euclidean'], ascending=False)
df
```

## Cosine

```python
def spatialDistance_cosine(vector1, vector2):
    return spatial.distance.cosine(vector1, vector2)
```

```python
df['score_cosine'] = df.apply(lambda x: spatialDistance_cosine(vector1, x['vector']), axis=1)
df = df.drop_duplicates(subset=['score_cosine']).sort_values(by=['score_cosine'], ascending=False)
df
```

## cityblock (Manhathan)

```python
def spatialDistance_cityblock(vector1, vector2):
    return spatial.distance.cityblock(vector1, vector2)
```
