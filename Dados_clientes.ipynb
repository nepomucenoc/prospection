## Geopy para codificar endereços

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
locator = Nominatim(user_agent="myGeocoder")
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
end['endereco_geolocalizado'] = end['completo'].apply(geocode)
```

```python
end.head()
```

```python
end['endereco_geolocalizado'].isna().sum()
```

```python
end['end_latitude'] = end['endereco_geolocalizado'].apply(lambda loc: loc.latitude if loc else None)
end['end_longitude'] = end['endereco_geolocalizado'].apply(lambda loc: loc.longitude if loc else None)
end['end_geocoder'] = end['endereco_geolocalizado'].apply(lambda loc: tuple(loc.point) if loc else None)
end.head()
```
