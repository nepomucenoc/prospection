```python
import http.client
import urllib.parse
import time


lista_json = []

for endereco in df['enderecos']:
    time.sleep(0.5)
    print(endereco)
    try:
      conn = http.client.HTTPConnection('geocode.xyz')
      params = urllib.parse.urlencode({
          'locate': endereco,
          'region': 'BR',
          'json': 1,
      })
      start_time = time.time()
      conn.request('GET', '/?{}'.format(params))
      res = conn.getresponse()
      end_time = time.time()
      print(f'Demora por request:{end_time-start_time}')
      data = res.read().decode('utf-8')
      print(data)
      result = json.loads(data)
      lista_json.append(result)
    except Exception as e:
      lista_json.append(np.nan)
      print(e)
      

lista_json
```
