## Distance matrix entre coordenadas geográficas

```python
import pandas as pd

import numpy as np
from math import sin, cos, sqrt, atan2, radians

from scipy.spatial.distance import pdist, squareform
```

```python
lat_long = pd.DataFrame({'LATITUDE': latitude, 'LONGITUDE': longitude})
```

```python
lat_long.head()
```

Referencia da coordenada 1 pois essa será o ponto em que o vendedor está, por exemplo. Será a primeira linha do dataframe.

```python
coord1 = eval(df.geocoder.loc[:0][0])
```

```python
def dist(coord1, coord2):
    """Function to compute the distance between two points x, y"""
    result = pd.DataFrame()

    lat1 = radians(coord1[0])
    lon1 = radians(coord1[1])
    lat2 = radians(coord2[0])
    lon2 = radians(coord2[1])

    R = 6373.0

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return round(distance, 4)
```

```python
df['distancia'] = df.geocoder.apply(lambda x: dist(coord1,eval(x)))
df.head()
```

```python
for distance in distances:
  if distance < 1:
    points = [f'point_{i}' for i in range(1, len(lat_long) + 1)]
    result = pd.DataFrame(squareform(distances), columns=points, index=points)

result
```

### Outra solução é:

O triângulo inferior da matriz de distância está vazio, pois a matriz é simétrica ( dist[i1,i2]==dist[i2,i1])

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos, sqrt, atan2, radians

lat_long = pd.DataFrame({'LATITUDE':latitude, 'LONGITUDE': longitude})
lat_long

test = lat_long.iloc[2:,:]

def distance(city1, city2):
    lat1 = radians(city1['LATITUDE'])
    lon1 = radians(city1['LONGITUDE'])
    lat2 = radians(city2['LATITUDE'])
    lon2 = radians(city2['LONGITUDE'])

    R = 6373.0

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance

dist = np.zeros([lat_long.shape[0],lat_long.shape[0]])
for i1, city1 in lat_long.iterrows():
    for i2, city2 in lat_long.iloc[i1+1:,:].iterrows():
        dist[i1,i2] = distance(city1, city2)

print(dist)
```
